```python
import numpy as np
import pandas as pd
```

```python
w = pd.read_csv('Data/top_500_matchs.csv', index_col = 0)
```

```python
w.head()
```

```python
rs_titles_artists = w['title_rs'] + "     " + w['artist_rs']
rs_titles_artists = list(rs_titles_artists.unique())

print('Recuperamos un total de ', len(rs_titles_artists), ' caniciones de las 500')
```

```python
rs_titles_artists = [{'title':val.split('     ')[0], 'artist':val.split('     ')[1]} for val in rs_titles_artists]
rs_titles_artists[:3]
```

```python
indices = []

for val in rs_titles_artists:
    
    ### Filter w to get all rows that match artist and title
    w_aux = w[w['title_rs'] == val['title']]
    w_aux = w_aux[w_aux['artist_rs'] == val['artist']]
    
    ### Retrieve the index of the highest scoring one
    # Note: There could be multiple qith the same score. We only take one
    indices.append(w_aux.sort_values(by = 'score').tail(1).index[0])
```

```python
w.loc[indices].sort_values(by = 'score', ascending = False).head(10)
```
